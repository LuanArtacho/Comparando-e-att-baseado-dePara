```python
import pandas as pd

caminho_baseUnificada = r'C:\Users\luanzera\Desktop\Estudos\Task BBA - Carteira RMs - PowerBi\BaseUnificada\BaseUnificada.xlsx'
caminho_dePara = r'C:\Users\luanzera\Desktop\Estudos\Task BBA - Carteira RMs - PowerBi\BaseUnificada\dePara.xlsx'

# Lendo os arquivos
df_base = pd.read_excel(caminho_baseUnificada)
df_de_para = pd.read_excel(caminho_dePara)

# Atualizar a coluna 'BANK_CITY_LENDER' com base na correspondência
for index, row in df_base.iterrows():
    banco_base = row['BANCO']
    for _, de_para_row in df_de_para.iterrows():
        banco_de_para = de_para_row['De']
        if banco_base == banco_de_para:
            df_base.at[index, 'BANK_CITY_LENDER'] = de_para_row['Para']
            break

# Exemplo de visualização dos dados atualizados
df_base

# Salvar o DataFrame atualizado em um novo arquivo Excel
# df_base.to_excel('caminho_do_arquivo_base_atualizado.xlsx', index=False)
```

```python
import pandas as pd

caminho_baseUnificada = r'C:\Users\luanzera\Desktop\Estudos\Task BBA - Carteira RMs - PowerBi\BaseUnificada\BaseUnificada.xlsx'
caminho_dePara = r'C:\Users\luanzera\Desktop\Estudos\Task BBA - Carteira RMs - PowerBi\BaseUnificada\dePara.xlsx'

# Lendo os arquivos
df_base = pd.read_excel(caminho_baseUnificada)
df_de_para = pd.read_excel(caminho_dePara)

# Criar um dicionário com a correspondência de valores
mapping_dict = dict(zip(df_de_para['De'], df_de_para['Para']))

# Atualizar a coluna 'BANK_CITY_LENDER' com base na correspondência
df_base['BANK_CITY_LENDER'] = df_base['BANCO'].map(mapping_dict).fillna(df_base['BANK_CITY_LENDER'])

# Filtrar o DataFrame original para conter apenas operações 'WK'
df_wk = df_base[df_base['TRANSACTION'] == 'WK']

# Remover o depara do DataFrame de WK
df_wk['BANK_CITY_LENDER'] = df_wk['BANCO']

# Salvar o DataFrame atualizado em um novo arquivo Excel
df_wk.to_excel('caminho_do_arquivo_base_wk_atualizado.xlsx', index=False)
```
